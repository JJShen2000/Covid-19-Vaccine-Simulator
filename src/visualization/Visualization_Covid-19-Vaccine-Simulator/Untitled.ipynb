```python
import csv
import pandas as pd
import numpy as np
```

```python
def create_toy_df(num_period, day_infect, day_tract):
    ''' Create the toy data.
    '''
    regions = []
    with open('census_tracts.csv', newline='', encoding="utf-8") as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            regions.append(str(row[0]).replace('臺', '台')+'/'+str(row[1]))

    df_base = pd.DataFrame(regions, columns=['Town'])
    df = df_base.copy()
    df['Period'] = np.ones((368,), dtype=int)
    df['num_infection'] = 0
    df.loc[0,'num_infection'] = day_infect
    
    for i in range(1, num_period):
        new_df = df_base.copy()
        new_df['Period'] = np.full((368, ), i+1)
        new_df['num_infection'] = 0
        for j in range(day_tract*(i+1)):
            new_df.loc[j,'num_infection'] = day_infect + day_infect*(i-j//day_tract)
            
        df = pd.concat([df, new_df])
    # df['num_infection'] = np.random.randint(0,30,size=368*num_period)
    df['num_dead'] = np.random.randint(0,5,size=368*num_period)
    df['ratio_infection'] = np.random.rand(368*num_period)
    df['ratio_dead'] = np.random.rand(368*num_period)
    return df
```

```python
toy_df = create_toy_df(num_period=10, day_infect=7, day_tract=20)
```

```python
toy_df.to_csv("toy_df1.csv", index=False)
```
